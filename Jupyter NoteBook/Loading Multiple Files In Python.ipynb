```python
import glob as glob
import pandas as pd
```

```python
# Define the path to the directory containing the CSV files
path = 'C:/Users/piett/Documents/GitHub/Ds_assignment#1/EuropeAirBnB/Weekdays/*.csv'
path2 = 'C:/Users/piett/Documents/GitHub/Ds_assignment#1/EuropeAirBnB/Weekends/*.csv'


# Load all the CSV files into a single dataframe
result_weekdays = pd.concat([pd.read_csv(file) for file in glob.glob(path)], ignore_index=True)
result_weekends = pd.concat([pd.read_csv(file) for file in glob.glob(path2)], ignore_index=True)

full_dataset= pd.concat([result_weekends,result_weekdays])
```
